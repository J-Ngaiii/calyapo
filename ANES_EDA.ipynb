```python
import pandas as pd
import numpy as np
import pyreadstat
import glob
import os
```

```python
anes_24_raw_path = 'data/raw/anes/anes-2024.csv'
cces_24_raw_path = 'data/raw/cces/cces-2024.csv'
igs_24_raw_path = 'data/raw/igs/igs-2024.csv'
rA24 = pd.read_csv(anes_24_raw_path) # check that it comes unweighted 
rC24 = pd.read_csv(cces_24_raw_path) # check that it comes unweighted 
rI24 = pd.read_csv(igs_24_raw_path)
```

```python
all_datapoints_count = 0

num_rA24 = len(rA24[rA24['V241529'] == 6])
num_rC24 = len(rC24[rC24['inputstate']==6])
num_rI24 = len(rI24)

all_datapoints_count += num_rA24 + num_rC24 + num_rI24
print(f"Number of datapoints in ANES: {num_rA24}")
print(f"Number of datapoints in CCES: {num_rC24}")
print(f"Number of datapoints in IGS: {num_rI24}")
```

```python
ppic_folder = 'data/raw/ppic/'

rPList24 = {}
num_rP24 = 0
for file_path in glob.glob(os.path.join(ppic_folder, "*.sav")):
    print(f"Opening {file_path}...")
    df, meta = pyreadstat.read_sav(file_path)
    rPList24[os.path.basename(file_path)] = { # extract file name with os.path.basename(file_path)
        'df':df, 
        'meta':meta
        }
    num_rP24 += len(df)
    print(f"Loaded {len(df)} rows, {len(df.columns)} columns")

all_datapoints_count += num_rP24
print(f"Total number of datapoints in PPIC: {num_rP24}")
```

```python
print(f"All datapoints so far: {all_datapoints_count}")
```
